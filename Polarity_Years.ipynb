```python
import pandas as pd
df_1 = pd.read_csv('preprocessed_data.csv')
df_1.head(3)
```

```python
## Add polarity scores to df_2019 - sentiment analysis 

from textblob import TextBlob

df_1['polarity'] = df_1['comments'].apply(lambda x: TextBlob(x).polarity)
df_1.head(3)
```

```python
df_19 = df_1[df_1["Year"]==2019]
df_20 = df_1[df_1["Year"]==2020]
df_21 = df_1[df_1["Year"]==2021]
df_22 = df_1[df_1["Year"]==2022]
```

```python
df_19_polar = df_19["polarity"].mean()
df_20_polar = df_20["polarity"].mean()
df_21_polar = df_21["polarity"].mean()
df_22_polar = df_22["polarity"].mean()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Year = ['2019', '2020', '2021', '2022']
Polarity = [df_19_polar, df_20_polar, df_21_polar, df_22_polar]
ax.bar(Year,Polarity)
plt.ylim([0.40, 0.42])
plt.xlabel("Years")
plt.ylabel("Polarity")
plt.show()
```
